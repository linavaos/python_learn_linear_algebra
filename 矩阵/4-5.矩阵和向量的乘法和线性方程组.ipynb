# 引列
* 在经济系统中投入，对IT、电子、破产、房产投入 $x_{it}、x_{e}、x_{m}、x_{h}$,存在如下方程组
$$

\begin{equation}
% 这是方程组大括号
\left\{
\begin{aligned}
x_{it} - 0.2x_e + 0.1x_m + 0.5x_h &= 100 \\
-0.5x_{it} - x_e + 0.2x_m + 0.1x_h &= 50 \\
-0.4x_e - x_m + 0.3x_h &= 20 \\
-0.2x_{it} + x_h &= 666
\end{aligned}
\right.
\end{equation}

$$

* 对应矩阵可以看成系数矩阵和x向量的点乘以
$$
\begin{equation}
\begin{pmatrix}
1 & -0.2 & 0.1 & 0.5 \\
-0.5 & -1 & 0.2 & 0.1 \\
0 & -0.4 & -1 & 0.3 \\
-0.2 & 0 & 0 & 1
\end{pmatrix}
\begin{pmatrix}
x_{it} \\
x_e \\
x_m \\
x_h
\end{pmatrix}
=
\begin{pmatrix}
100 \\
50 \\
20 \\
666
\end{pmatrix}
\end{equation}

$$

* 也就那矩阵的行向量和x向量的列向量的点乘


* 推论：当一个m行*n列的矩阵和一个列向量相乘。则合法的计算必须
    * 列向量行数必须等于n
    * 且相乘的结果为一个新的列向量。列向量的个数等于m
