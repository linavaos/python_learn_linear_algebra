# 矩阵乘法
可以看成矩阵和一个列向量的点乘，最后结果还是一个矩阵。



啥意思呢？
就是下面公式，就是矩阵点乘等于第一个矩阵点乘第二个矩阵没一列的向量,假设我们下标表示取行，上标表示取列
$$
T \cdot P
=
\begin{pmatrix}
  T \cdot \vec{p^{(0)}}
, T \cdot \vec{p^{(1)}}
，T \cdot \vec{p^{(2)}}
\end{pmatrix} 
\\
=
\begin{pmatrix}
  \vec{t_0} \cdot \vec{p^{(0)}}
  ,\vec{t_0} \cdot \vec{p^{(1)}}
   ,\vec{t_0} \cdot \vec{p^{(2)}}
   \\
   \vec{t_1} \cdot \vec{p^{(0)}}
  ,\vec{t_1} \cdot \vec{p^{(1)}}
   ,\vec{t_1} \cdot \vec{p^{(2)}}
\end{pmatrix} 

$$

* 也就是，矩阵相乘实际是T矩阵的行向量和P矩阵的列向量点乘，所以要保证T矩阵的列等于P矩阵的行。也就是存在一个矩阵 r*k和存在一个矩阵k*c 点乘。结果会生产一个r*c的矩阵。同时
$$
T \cdot P \neq P \cdot C
$$
