# 单位矩阵
1. 主对角线上的元素全部为1的矩阵。
2. 既然要求对角线上的元素全部为1。那么行不等于列的矩阵天生就有元素不在对角线上，所以必须是方正

3. 在明确点就是某个元素的行索引=列索引，为1的，其他位置为0的时候就是单位矩阵。

# 单位矩阵点乘性质
* 有一个单位矩阵I和一个单位矩阵J,有一个矩阵A（注意A未必是方阵）,满足
$$
I\cdot A = A \cdot J = A
$$                                                                                                                         

* 但是
$$
I \neq J
$$
