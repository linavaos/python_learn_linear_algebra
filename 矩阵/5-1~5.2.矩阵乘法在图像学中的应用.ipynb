## 上下翻转
 


* 如矩阵组成F要切换翻转的F如何操作，其本质就是求P
$$
P \cdot
\begin{pmatrix}
 x
 \\
 y
\end{pmatrix}
=
\begin{pmatrix}
 x
 \\
 -y
\end{pmatrix}
$$
* 根据向量乘法，我们知道P的列=后面矩阵的行=2，最后结果为2行，所以p的行=2。我们s


$$
\begin{pmatrix}
a,b
\\
c,d
\end{pmatrix}
 \cdot
\begin{pmatrix}
 x
 \\
 y
\end{pmatrix}
=
\begin{pmatrix}
 x
 \\
 -y
\end{pmatrix}
$$


* 可得
$$
ax+by=x
\\
cx+dy=-y
$$


* a=1,b=0,c=0,d=-1

* 也就是上下翻转
\begin{pmatrix}
 1，0
 \\
 0，-1
\end{pmatrix}

* 也可以的左右翻转为
\begin{pmatrix}
 -1，0
 \\
 0，1
\end{pmatrix}

* 原点翻转
\begin{pmatrix}
 -1，0
 \\
 0，-1
\end{pmatrix}

# 水平错切
* 就是对x进行变化的得到x'=x+ay而y不变,最终实现效果是类似于一个字体的斜体


* 其实就是F中的某个点，其实要移动的距离其实可以看成上图矩形变成平行四边形的过程是一样的。都是x变化，y轴不变化。那么x如何变化的，就是
$$
x'=x + tan\alpha *y,之里的tan\alpha当中一个常量a。来控制倾斜角度
$$


* 所以就是求
$$
T \cdot \begin{pmatrix} x \\ y \end{pmatrix} =
\begin{pmatrix}
x + ay \\
y
\end{pmatrix}

$$


* 设T为
$$
T=
\begin{pmatrix}
a,b
\\
c,d
\end{pmatrix}
$$

* 安之前点乘，建立方程可以得到
$$
T=
\begin{pmatrix}
-1,a
\\
0,1
\end{pmatrix}
$$

* 如果a=0,说明$tan\alpha=0$,说明倾斜0度$，加上-1,说明x水平翻转

## 旋转


* 如上图，图中的F可以看出关于x和y组成的矩阵，想在F旋转$\theta$后，得到新的矩阵。这个过程是如何变化的（一矩阵点乘的角度看）。也就求T
$$
T \cdot
\begin{pmatrix}
 x
 \\
 y
\end{pmatrix}
=
\begin{pmatrix}
x^\prime
 \\
y^\prime
\end{pmatrix}
$$

* 上面的式子在平面坐标系就是向量的旋转

* 我们把（x,y）向量 到直角坐标系的夹角为$\alpha$,把(x,y)向量的模设为d
$$
 d=x*cos(\alpha)=x\prime*cos(\alpha-\theta)
$$


$$
 \text{可得：}
 x\prime=\frac{cos(\alpha-\theta)}{cos(\alpha)}*x
$$

$$
 \text{同理，可得：}
y\prime=\frac{sin(\alpha-\theta)}{sin(\alpha)}*y
$$

$$
\text{余弦两角差值公式} 
cos(\alpha-\theta)=cos\alpha*cos\theta+sin\alpha*sin\theta
\\
\text{正弦两角差值公式}
sin(\alpha-\theta)=sin\alpha*cos\theta-cos\alpha*sin\theta
$$

$$
\text{所以}
x\prime=\frac{cos\alpha*cos\theta+sin\alpha*sin\theta}{cos\alpha}
\\
=cos\theta*x+sin\theta*\frac{sin\alpha}{cos\alpha}*x
\\
=
cos\theta*x+sin\theta*\tan\alpha*x
$$

$$
由于 tan\alpha=\frac{y}{x}
$$

$$
x\prime = cos\theta*x+sin\theta*y
$$

$$
同理可得
y\prime = cos\theta*y-sin\theta*x
$$

$$ 
\text{所以转为求这个式子种的T:}
\\
T \cdot \begin{pmatrix} x \\ y \end{pmatrix} =
\begin{pmatrix}
\cos\theta \cdot x + \sin\theta \cdot y \\
-\sin\theta \cdot x + \cos\theta \cdot y
\end{pmatrix}

$$

$$
\text{设:}
T=\begin{pmatrix} 
a,b
\\
c,d
\end{pmatrix} 
$$

$$
\text{可得}：
T=\begin{pmatrix} 
cos\theta,sin\theta
\\
-sin\theta,cos\theta
\end{pmatrix} 
$$
