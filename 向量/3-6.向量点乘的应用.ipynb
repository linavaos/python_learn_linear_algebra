# 推荐系统


* 通过计算$\theta$,我们可以知道是/theta是是锐角还是钝角$

$$
\cos \theta  = \frac{{\vec{u}} \cdot \vec{v}}{\|\vec{u}\| \|\vec{v}\| }  
$$

* 当$\cos \theta<0$：说明具有一定相似性，方向上相似
* 当$\cos \theta=0$：说明不具备相似性
* 当$\cos \theta>0$：说明两个向量相互背离


# 计算投影
* 计算下图中投影坐标位置
* 

* 距离d，也是（0,0）画圈位置的模
$$d=\|\vec{v}\|*\cos\theta$$

* 而
$$
\cos \theta  = \frac{{\vec{u}} \cdot \vec{v}}{\|\vec{u}\| \|\vec{v}\| }  
$$

* 所以
$$d=\frac{{\vec{u}} \cdot \vec{v}}{ \|\vec{v}\| }  $$

* 计算u向量的单位向量$\hat{u}$

* 之后通过投影向量p
$$\vec{p}=d*\hat{u}$$
