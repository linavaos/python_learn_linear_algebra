# 什么是向量
* 就是一组数的表示方法
* 向量（Vector）是线性代数研究的基本原始

# 应用场景
* 如一个向量只二个数，用来描述经纬度、平面坐标中的位置、水平速度（大小、方向）
* 如一个向量三个数，用来描述三维空间位置
* 大于三个数，通常是多维，从几何层面看比较抽象，但是不影响代数分析，如描述一个房子


# 向量和标量代数表示形式
* 代数就是有字母代替数，就是我们认为的变量
* 标量就是一个数，变量表示如v
* 向量变量的表示 $\vec{v}$
* 在几何中我们想表示O到A的向量$\vec{OA}$


# 行向量和列向量
* 行向量:(3,4)
* 列向量:$\begin{pmatrix}3, \\ 4\end{pmatrix}$ 或者表示为$(3,4)^T$
