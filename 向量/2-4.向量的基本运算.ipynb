# 向量加法

## 公式
$$
\
\begin{pmatrix}  
v_1 \\
v_2 \\
\vdots \\
v_n  
\end{pmatrix}  
+  
\begin{pmatrix}  
u_1 \\
u_2 \\
\vdots \\
u_n  
\end{pmatrix}  
=  
\begin{pmatrix}  
v_1 + u_1 \\
v_2 + u_2 \\
\vdots \\
v_n + u_n  
\end{pmatrix}  
\
$$

## 几何意义

$$(5,2)^T+(2,5)^T=?$$

* 如下图是（5，2）和 （2，5）向量在直角坐标系的位置


* 两个向量相加，相当于先从(0,0) 位移（5，2）移动到（5，2）再从 (5,2) 位移 (2,5) 到 （7，7）

* 这个向量的移动和（0，0）移动到（7，7）是等价的


* 也就是平行四边行的对角线


* 所以向量加法的集合意义是两个向量作为边的评选四边形对角线向量

v_n

k \cdot v_1 \\
k \cdot v_2 \\
\vdots \\

\end{pmatrix}

$$
