# 两个向量点乘

$$
\vec{u} \cdot \vec{v} 
= u_1 v_1 + u_2 v_2 + \dots + u_n v_n
= \sum_{i=1}^{n} (u^{(i)}v^{(i)})
= \|\bar{u}\| \|\bar{v}\| \cos \theta  

$$

* 两个向量点乘，也叫两个向量的内积
* 其结果是一个标量

# 二维向量点乘几何公式推论
$$

{\vec{u}} \cdot \vec{v} = x_1 \cdot x_2 + y_1 \cdot y_2 = \|\vec{u}\| \|\vec{v}\| \cos \theta  
$$

* 暂时只证明
$$
{\vec{u}} \cdot \vec{v} = \|\vec{u}\| \|\vec{v}\| \cos \theta  
$$

* 首先我们需要明确向量减法的几何意义。没错就是两个向量夹角的对边
    * 如下图利用向量加法可以得出   
    * $\vec{v}+(\vec{u}-\vec{v}) = \vec{u}$  


* 另$\vec{u}=u,\vec{v}=v,\vec{j}=\vec{u}-\vec{v}$,主要取模|看的眼花

* 利用三角函数余弦定理可得


$$
j^2=u^2+v^2-2*u*v*\cos \theta  （余弦定理）
$$

* 变化的
$$
u*v*\cos \theta=\frac{1}{2}(u^2+v^2-j^2) 
$$

* 带入$\vec{u}=(x_1,y_1),\vec{v}=(x_2,y_2),\vec{j}=(x_1-x_2,y_1-y_2)$ 
$$
u*v*\cos \theta=\frac{1}{2}(x_1^2+y_1^2 + x_2^2+y_2^2-[(x_1-x_2)+(y_1-y_2)]^2) 
$$

* 整理后可证
$$
u*v*\cos \theta=x_1*x_2+y_1*y_2
$$
