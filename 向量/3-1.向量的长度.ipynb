# 向量的长度（模）


# 单位向量
* 就是那向量的每个分量（元素）除以向量的模
$$
\begin{align*}  
\bar{u} & = (u_1, u_2, \ldots, u_n)^T \\
\hat{u} & = \frac{1}{\|\bar{u}\|} \bar{u} = \left( \frac{u_1}{\|\bar{u}\|}, \frac{u_2}{\|\bar{u}\|}, \ldots, \frac{u_n}{\|\bar{u}\|} \right)  
\end{align*}
$$

* 其实比如(3,4)向量的模是5。现在要求模是1的时候，每个分量是多少。而单位向量是表示的方向。

* 根据$\bar{u}$求$\hat{u}$的过程，叫做归一化、规范化

# 标准单位向量
* 就是只有一个分量为1。其他分量为0的向量
